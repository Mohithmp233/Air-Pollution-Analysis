```python
file_path = 'city_day.csv'
data = pd.read_csv(file_path)
pollutants = ['PM2.5', 'PM10', 'NO2', 'CO', 'SO2', 'O3']
data_cleaned = data[pollutants].dropna()
def visualize_air_quality_distribution_stats(data_cleaned, pollutants):
    while True:
        print("\nAvailable pollutants to visualize:")
        for idx, pollutant in enumerate(pollutants, start=1):
            print(f"{idx}. {pollutant}")

        try:
            choice_pollutant = int(input(f"\nEnter the number corresponding to the pollutant you want to visualize (1-{len(pollutants)}): "))
            if choice_pollutant < 1 or choice_pollutant > len(pollutants):
                print("Invalid choice. Please select a valid number.")
                continue

            selected_pollutant = pollutants[choice_pollutant - 1]
            print("\nAvailable statistics:")
            print("1. Mean")
            print("2. Median")
            print("3. Maximum")
            print("4. Minimum")

            choice_stat = int(input(f"Enter the number corresponding to the statistic you want to use (1-5): "))
            if choice_stat < 1 or choice_stat > 5:
                print("Invalid choice. Please select a valid number.")
                continue

            stat_functions = {
                1: 'mean',
                2: 'median',
                3: 'max',
                4: 'min'
            }
            stat_func = stat_functions[choice_stat]
            stat_value = getattr(data_cleaned[selected_pollutant], stat_func)()
            print(f"\n{stat_func.capitalize()} of {selected_pollutant}: {stat_value}")
            
            # Visualization
            plt.figure(figsize=(10, 6))
            sns.histplot(data_cleaned[selected_pollutant], bins=30, kde=True, color='blue')
            plt.title(f'Distribution of {selected_pollutant} (Statistic: {stat_func.capitalize()})', fontsize=14)
            plt.xlabel(f'{selected_pollutant} Levels', fontsize=12)
            plt.ylabel('Frequency', fontsize=12)
            plt.tight_layout()
            plt.show()

        except ValueError:
            print("Invalid input. Please enter a valid number.")

        # Ask the user if they want to continue or exit
        continue_choice = input("\nDo you want to analyze another pollutant? (yes/no): ").strip().lower()
        if continue_choice != 'yes':
            print("Exiting the program. Goodbye!")
            break

# Call the function
visualize_air_quality_distribution_stats(data_cleaned, pollutants)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'city_day.csv'
data = pd.read_csv(file_path)

# Ensure 'Date' column is in datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Select 5 major cities
major_cities = ['Delhi', 'Mumbai', 'Chennai', 'Kolkata', 'Bengaluru']

# Create a new column to categorize cities into 'Other' if they are not in the major cities list
data['City_Category'] = data['City'].apply(lambda x: x if x in major_cities else 'Other')

# Create a line plot for AQI trends
plt.figure(figsize=(12, 8))

# Plot AQI trend for major cities
for city in major_cities:
    city_data = data[data['City_Category'] == city].sort_values(by='Date')
    plt.plot(city_data['Date'], city_data['AQI'], marker='o', label=city)

# Plot AQI trend for other cities combined as 'Other'
other_data = data[data['City_Category'] == 'Other'].groupby('Date')['AQI'].mean().reset_index()
plt.plot(other_data['Date'], other_data['AQI'], marker='x', label='Other', color='gray')

# Add title, labels, and legend
plt.title('AQI Trends Over Time for Major Cities and Others', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('AQI', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Cities', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'city_day.csv'  # Make sure the file path is correct
data = pd.read_csv(file_path)

# Ensure 'Date' column is in datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Define AQI categories (adjust these ranges based on your classification system)
bins = [0, 50, 100, 150, 200, 300, 500]  # Define AQI category ranges
labels = ['Moderate', 'Satisfactory', 'Poor', 'Very Poor', 'Good', 'Severe']

# Create a new column for AQI bucket category based on the AQI values
data['AQI_Category'] = pd.cut(data['AQI'], bins=bins, labels=labels, right=False)

# Function to check cities in a particular AQI bucket
def cities_in_aqi_bucket(aqi_bucket):
    # Filter the data for the specified AQI bucket
    cities_in_bucket = data[data['AQI_Category'] == aqi_bucket]['City'].value_counts()
    return cities_in_bucket

# Loop to ask for AQI bucket selection repeatedly
while True:
    print("\nAvailable AQI Categories:")
    for i, label in enumerate(labels, 1):  # Start enumeration from 1
        print(f"{i}. {label}")
    
    # Get user input for AQI category number
    try:
        choice = int(input("\nEnter the number corresponding to the AQI bucket to see the cities (or 0 to exit): "))
        
        if choice == 0:
            print("Exiting the program.")
            break  # Exit the loop if the user chooses 0
        
        if 1 <= choice <= len(labels):
            aqi_bucket = labels[choice - 1]  # Get the AQI category based on the user's choice
            cities = cities_in_aqi_bucket(aqi_bucket)
            
            # Display the cities
            print(f"\nCities in the '{aqi_bucket}' AQI bucket: {cities.index.tolist()}")
            
            # Plotting the cities in the selected AQI bucket
            plt.figure(figsize=(10, 6))
            sns.barplot(x=cities.index, y=cities.values, palette="viridis")
            plt.title(f'Cities in the "{aqi_bucket}" AQI Bucket', fontsize=16)
            plt.xlabel('Cities', fontsize=12)
            plt.ylabel('Frequency', fontsize=12)
            plt.xticks(rotation=45)
            plt.tight_layout()
            plt.show()
        
        else:
            print(f"Invalid choice. Please enter a number between 1 and {len(labels)} or 0 to exit.")
    except ValueError:
        print("Invalid input. Please enter a valid number.")
```
